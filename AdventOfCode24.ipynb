```python
import numpy as np
import sys
import enum

class instruction():
    def __init__(self, instruction, argument1, argument2):
        self.instruction = instruction
        self.argument1 = argument1
        self.argument2 = argument2
        
    def __str__(self):
        return f"{self.instruction}|{self.argument1}|{self.argument2}"

class instructionset():
    def __init__(self, instructions):
        self.instructions = instructions
    def __str__(self):
        return f"{self.instructions}"
    
    def Execute(self, input, printValue = False):
        datastore = [0,0,0,0]
        inputIndex= 0
        dataAddress = {}
        dataAddress['E'] =5000
        dataAddress['w'] = 0
        dataAddress['x'] = 1
        dataAddress['y'] = 2
        dataAddress['z'] = 3

        for instruction in self.instructions:

            

            

            # switch on instruction
            if instruction.instruction == 'inp':
                if printValue: print(inputIndex, 'last input:', input[inputIndex-1] , 'z Value:', datastore[dataAddress['z']])
                datastore[dataAddress[instruction.argument1]] = input[inputIndex]
                inputIndex += 1
            elif instruction.instruction == 'add':
                if not instruction.argument2.isalpha():
                    datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] + int(instruction.argument2)
                else:
                    datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] + datastore[dataAddress[instruction.argument2]]
            elif instruction.instruction == 'mul':
                if not instruction.argument2.isalpha():
                    datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] * int(instruction.argument2)
                else:
                    datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] * datastore[dataAddress[instruction.argument2]]
            elif instruction.instruction == 'mod':
                if not instruction.argument2.isalpha():
                    if ( int(instruction.argument2) == 0):
                        # trow error
                        print("Error: division by zero")
                        break
                    if not instruction.argument2.isalpha():
                        datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] % int(instruction.argument2)
                else :
                    if ( datastore[dataAddress[instruction.argument2]] == 0):
                        # trow error
                        print("Error: division by zero")
                        break
                    else:
                        datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] % datastore[dataAddress[instruction.argument2]]
            elif instruction.instruction == 'div':
                if not instruction.argument2.isalpha():
                    if ( int(instruction.argument2) == 0):
                        # trow error
                        print("Error: division by zero")
                        break
                    if not instruction.argument2.isalpha():
                        datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] // int(instruction.argument2)
                else :
                    if ( datastore[dataAddress[instruction.argument2]] == 0):
                        # trow error
                        print("Error: division by zero")
                        break
                    else:
                        datastore[dataAddress[instruction.argument1]] = datastore[dataAddress[instruction.argument1]] // datastore[dataAddress[instruction.argument2]]
                
            elif instruction.instruction == 'eql':
                if printValue: print(f'{inputIndex}: {instruction}, {datastore}')
                if (instruction.argument2.isnumeric()):
                    if (datastore[dataAddress[instruction.argument1]] == int(instruction.argument2)):
                        datastore[dataAddress[instruction.argument1]] = 1
                    else:
                        datastore[dataAddress[instruction.argument1]] = 0
                else :
                    if (datastore[dataAddress[instruction.argument1]] == datastore[dataAddress[instruction.argument2]]):
                        datastore[dataAddress[instruction.argument1]] = 1
                    else:
                        datastore[dataAddress[instruction.argument1]] = 0
        if printValue: print(inputIndex,'last input:', input[inputIndex-1] , 'z Value:', datastore[dataAddress['z']])
        return datastore[dataAddress['z']]


def GetInstructions(dataset):
    data = []
    with open(dataset, 'r') as f:
        data = f.readlines()

    
    # strip the new line character
    data = [x.strip() for x in data]
    
    # split spaces 
    data = [x.split(' ') for x in data]
    instructions = []
    for i in range(len(data)):
        if (len(data[i])==3):
            instructions.append(instruction(data[i][0], data[i][1], data[i][2]))
            #print(instructions[-1])
        else:
            instructions.append(instruction(data[i][0], data[i][1], 'E'))
            #print(instructions[-1])
    alu = instructionset(instructions)

    return alu


    
```
