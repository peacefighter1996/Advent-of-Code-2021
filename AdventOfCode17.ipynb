```python
import numpy as np
import cv2
import math
import sys
import time
import enum

# Get the risk area
def GetTaget(dataset):
    data = []
    # target area: x=240..292, y=-90..-57
    with open(dataset, 'r') as f:
        data = f.readlines()
    # strip the new line character
    data = [x.split(':') for x in data]

    data = [x[1].split(',') for x in data]
    data = [[x[0].split('=')[1].split('..'), x[1].split('=')[1].split('..')] for x in data]
    data = [[int(x[0][0]), int(x[0][1]), int(x[1][0]), int(x[1][1])] for x in data]
    return data

def TriagleNumber(n):
    return int(n*(n+1)/2)

def GetMaximumY(target):
    xRange = [target[0], target[1]]
    yRange = [target[2], target[3]]
    results = []
    for i in range(max(xRange)+1):
        print('i = ', i)
        for j in range(i,-1,-1):
            postionx = TriagleNumber(i)-TriagleNumber(j)
            steps = i-j
            #print('postionx = ', postionx)
            #print(f'x postion: {postionx},  {i} {j}, {postionx <= max(xRange) and postionx >= min(xRange)}')
            if postionx <= max(xRange) and postionx >= min(xRange):
                initaialVelosityx = i
                print(min(yRange))
                breakfull = False
                
                if j == 0:
                    for initaialVelosityY in range(abs(max(target)), 0, -1):
                        postiony = 0
                        for l in range( 0, abs(max(target))+1, 1):
                            postiony = TriagleNumber(initaialVelosityY)-TriagleNumber(initaialVelosityY+l)
                            if postiony >= min(yRange) and postiony <= max(yRange):
                                print(f'i: {i}, j: {j}, initialVelosityY:{initaialVelosityY}, l: {l}, postion: ({postionx},{postiony}) height: {TriagleNumber(initaialVelosityY)}')
                                results.append([initaialVelosityx, initaialVelosityY])
                                break
                            if postiony < min(yRange):
                                break
                
                for initaialVelosityY  in range(abs(max(target)), min(yRange)-1, -1):
                    xv = initaialVelosityx
                    y = initaialVelosityY
                    postiony = 0
                    postionx = 0
                    for step in range(0, steps):
                        postiony += y
                        y += -1
                        postionx += xv
                        xv = max(xv-1, 0)
                        if postiony >= min(yRange) and postiony <= max(yRange) and postionx <= max(xRange) and postionx >= min(xRange):
                            print(f'i: {i}, j: {j}, initialVelosityY:{initaialVelosityY}, step: {step+1}, postion: ({postionx},{postiony})')
                            results.append([initaialVelosityx,initaialVelosityY ])
                            break
                        if postiony < min(yRange):
                            break
                        
                    
            if(postionx > max(xRange)):
                break
    # Remove the duplicate
    newResults = []
    for i in range(len(results)):
        found = False
        for j in range(len(newResults)):
            if results[i][0] == newResults[j][0] and results[i][1] == newResults[j][1]:
                found = True
                break
        if not found:
            newResults.append(results[i])
    
    return newResults

# Brute force

def GetResults(target):
    maxvalue = max(target)
    rangex= [target[0], target[1]]
    rangey = [target[2], target[3]]
    results = []
    minimumx = 0
    for i in range(min(rangex)):
        postionx = TriagleNumber(i)
        if (postionx >= min(rangey) and postionx <= max(rangey)):
            minimumx = i

    for Initialx in range(1, maxvalue+1, 1):
        for Initialy in range(-maxvalue, maxvalue+1, 1):
            y = Initialy
            x = Initialx
            postionx = 0
            postiony = 0
            #print(f'Initialx: {Initialx}, Initialy: {Initialy}')
            while(True):
                postionx += x
                x = max(x-1, 0)
                postiony += y
                y += -1

                #print(f'x: {postionx}, y: {postiony}')
                if postionx >= min(rangex) and postionx <= max(rangex) and postiony >= min(rangey) and postiony <= max(rangey):
                    #print(f'i: {Initialx}, j: {Initialy}, step: {step+1}, postion: ({postionx},{postiony})')
                    results.append([Initialx, Initialy])
                    break
                if postiony < min(rangey):
                    break
                if x == 0 and not(postionx >= min(rangex) and postionx <= max(rangex)) or postionx > max(rangex):
                    break
    return results
    
```
