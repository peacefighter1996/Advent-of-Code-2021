```python
import numpy as np
import cv2

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Fold:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def performFold(self, Image):
        # cut the image on the fold
        if self.x == 0:
            slice1 = Image[:, :self.y]
            slice2 = Image[:, self.y+1:]
            slice2 = np.flip(slice2, 1) 
            print(slice1.shape, slice2.shape)
            slice1+= slice2
            return slice1
        elif self.y == 0:
            slice1 = Image[:self.x, :]
            slice2 = Image[self.x+1:, :]
            print(slice1.shape, slice2.shape)
            slice2 = np.flip(slice2, 0) 
            return slice1 + slice2             
    
    def __str__(self):
        return "Fold: y: {}, x: {}".format(self.x, self.y)
# Get the area of the image of from the document
def GetImageAndFolds(dataset):
    data = []
    with open(dataset, 'r') as f:
        data = f.readlines()
    # strip the new line character
    data = [x.strip() for x in data]
    
    # get all points
    points = []
    i = 0
    xMax = 0
    yMax = 0
    for i in range(len(data)):
        if data[i] == '':
            break
        else:
            x, y = data[i].split(',')
            points.append(Point(int(x), int(y)))
            if int(x) > xMax:
                xMax = int(x)
            if int(y) > yMax:
                yMax = int(y)
    
    # Add all point to Image
    Image = np.zeros((yMax+3, xMax+1), dtype=np.uint8)
    for point in points:
        Image[point.y, point.x] = 1
    
    # get all folds
    folds = []
    for i in range(i+1, len(data)):
        if data[i] == '':
            break
        else:
            _, _ , Item = data[i].split(' ')
            Item = Item.split('=')
            if Item[0] == 'x':
                folds.append(Fold(0, int(Item[1])))
            else:
                folds.append(Fold(int(Item[1]), 0 ))
    
    return Image, folds


            
```
