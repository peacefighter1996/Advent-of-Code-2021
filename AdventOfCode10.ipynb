```python
import numpy as np

# Get the area of the image of from the document
def GetCodeStack(dataset):
    data = []
    with open(dataset, 'r') as f:
        data = f.readlines()
    # strip the new line character
    data = [x.strip() for x in data]

    

    return data

data = GetCodeStack('./data/aoc10.txt')

combinations = [['(', ')', 3, 1], ['[', ']', 57, 2], ['{', '}', 1197, 3], ['<', '>', 25137, 4]]
corrupted = []
syntaxErrorScore = 0
syntaxComplitionScores = []

for i in range(len(data)):
    stack = []
    breakbit = False
    for j in range(len(data[i])):
        for k in range(len(combinations)):
            if data[i][j] == combinations[k][0]:
                stack.append(combinations[k][0])
            elif data[i][j] == combinations[k][1]:
                if stack[-1] == combinations[k][0]:
                    stack.pop()
                else:
                    #print(f'expected a cobination with {stack[-1]}, got {combinations[k][1]}')
                    #print(f'Corrupt data: {data[i]}')
                    breakbit = True
                    syntaxErrorScore += combinations[k][2]
                    corrupted.append(data[i])
                    break
        if breakbit:
            break
    # part 2 - Get Completion score
    if not(breakbit):
        syntaxCompletionRowScore = 0
        stack.reverse()
        
        for i in range(len(stack)):
            for j in range(len(combinations)):
                if stack[i] == combinations[j][0]:
                    syntaxCompletionRowScore = syntaxCompletionRowScore * 5 + combinations[j][3]
        print(f'stack: {stack}')
        print(f'syntaxCompletionScore {syntaxCompletionRowScore}')
        syntaxComplitionScores.append(syntaxCompletionRowScore)

syntaxComplitionScores.sort()
print(f'Score Part 1: {syntaxErrorScore}')
print(f'Score Part 2: {len(syntaxComplitionScores)}')
print(f'Score Part 2: {syntaxComplitionScores[int(len(syntaxComplitionScores)/2)]}')
```
